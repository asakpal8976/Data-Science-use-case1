```python
data = pd.read_csv('Analysis_data.csv')
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
c_data = data.copy()
for columns in c_data.select_dtypes(object):
    c_data[columns] = enc.fit_transform(c_data[columns])

c_data.head()

X,Y = c_data.drop(columns = ['Loan_Status']),c_data['Loan_Status']
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
naive = MultinomialNB()
```

```python
naive.fit(X,Y)
```

```python
p = X.iloc[1,:-1]
p
```

```python
naive.predict_proba
```
