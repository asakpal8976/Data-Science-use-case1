```python
import pandas as pd
import numpy as np

data = [1,'1',np.nan, 'a', 'b', np.nan]
#x = []

print(data)
(pd.Series(data).isnull().sum()/len(data))*100
```

```python
# add 100 items in this list, if the number is even add the number, 
#if the number is multiple of 3 add the square otherwise add null value and calculate the null %

list = []
for i in range(100):
    if i%2 == 0:
        list.append(i)
    elif i%3 == 0:
        list.append(i**2)
    else:
        list.append(np.nan)

print(list)
(pd.Series(list).isnull().sum()/len(list))*100
```

```python
#write a pandas program to compare the element of 2 pandas series
serie1 = [2,4,4,8,10]
serie2 = [1,3,5,7,9]

for i , j in zip(serie1,serie2):
    print(i,j)
    if [i] > [j]:
        print(f'{i} is greater than {j}')
    elif [i] < [j]:
        print(f'{j} is greater than {i}')
    else:
        print(f'Both the series are equal')
```

```python
#create some dummy data
df1 = pd.DataFrame({
    'A' : ['A0','A1','A2','A3'],
    'B' : ['B0','B1','B2','B3'],
},   index = ['K0','K1','K2','K3'])

df2 = pd.DataFrame({
    'C' : ['C0','C1','C2'],
    'D' : ['D0','D1','D2'],
},   index = ['K0','K1','K4'])

df3 = pd.DataFrame({
    'E' : ['E0','E1','E2','E3'],
    'F' : ['F0','F1','F2','F3'],
},  index = ['K0','K1','K5','K6'])

#inner join
inner_join = df1.join(df2 , how = 'inner')
print('\nInner Join :')
print(inner_join)

#outer join
outer_join = df1.join(df2, how = 'outer')
print('\nOuter Join :')
print(outer_join)

#left join
left_join = df1.join(df3, how = 'left')
print('\nleft join :')
print(left_join)

#right join
right_join = df1.join(df3, how = 'right')
print('\nright join :')
print(right_join)
```
